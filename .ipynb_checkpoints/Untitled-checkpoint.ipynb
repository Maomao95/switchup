```python
bootcamp = pd.read_csv('description.csv')
bootcamp.columns =['honor','bootcamp_name','average_score','review_number','price','locations','programs_available','scholarships']
review = pd.read_csv('reviews.csv')
```

```python
review.review_content = review.review_content.apply(lambda x:x[26:][:-22])
review.review_content = review.review_content.apply(lambda x:BeautifulSoup(x).get_text())
review.review_content = review.review_content.apply(lambda x: re.sub('\\\\n','',x))
review.review_content = review.review_content.apply(lambda x: re.sub('\\\\"','',x))
review.review_title = review.review_title.apply(lambda x: re.sub('\"','',x))
review.date = pd.to_datetime(review.date)
```

```python
score_performance = bootcamp.average_score.dropna()
plt.figure(figsize=(12,5))
sns.distplot(score_performance[score_performance.values >70],fit = norm)
fig = plt.figure()
```

```python
bootcamp2 = bootcamp.dropna(subset=['programs_available'])
bootcamp2.programs_available = bootcamp2.programs_available.apply(lambda x: re.split(', and |,| and |  ',x))
def unlist(row, col, new_col):
    col_items = row[col]
    num_items = len(col_items)
    rows = [{**row, new_col: col_items[i]}for i in range(num_items)]
    return rows
new_rows = bootcamp2.apply(lambda row: unlist(row,'programs_available', 'category'), 1)
new_rws = [rw for row in new_rows for rw in row]
df = pd.DataFrame(new_rws)
df.category = df.category.apply(lambda x: x.strip())
combine = pd.merge(df,review,on = 'bootcamp_name')
group= combine.groupby(['category','date'])
group2= combine.groupby('category')
top_5 = group2.size().sort_values(ascending=False)[:5].index.to_list()
group_mean = group.mean().reset_index('date')
top_5_category = group_mean[group_mean.index.isin(top_5)]
top_5_category['year'] = top_5_category.date.dt.year
group_by_year = top_5_category.groupby(['category','year']).mean()
group_by_year = group_by_year.reset_index('category')
group_by_year = group_by_year.drop([2013,2014,2015])
```

```python
import plotly.plotly
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
plotly.offline.init_notebook_mode(connected=True)

date = group_by_year.index
Coding_Bootcamps = group_by_year[group_by_year.category =='Coding Bootcamps'].overall_score.values
Full_Stack_Web_Development = group_by_year[group_by_year.category =='Full-Stack Web Development'].overall_score.values
Online_Courses = group_by_year[group_by_year.category =='Online Courses'].overall_score.values
Data_Science = group_by_year[group_by_year.category =='Data Science'].overall_score.values
Web_Design= group_by_year[group_by_year.category =='Web Design (UX/UI)'].overall_score.values

# Create and style traces
trace0 = go.Scatter(
    x = date,
    y = Coding_Bootcamps,
    name = 'Coding Bootcamps',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        )
)
trace1 = go.Scatter(
    x = date,
    y = Full_Stack_Web_Development,
    name = 'Full-Stack Web Development',
    line = dict(
        color = ('rgb(22, 96, 167)'))
)
trace2 = go.Scatter(
    x = date,
    y = Online_Courses,
    name = 'Online Courses',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        dash = 'dash') # dash options include 'dash', 'dot', and 'dashdot'
)
trace3 = go.Scatter(
    x = date,
    y = Data_Science,
    name = 'Data Science',
    mode = 'marker',
    line = dict(
        color = ('rgb(22, 96, 167)'),
        dash = 'dash')
)
trace4 = go.Scatter(
    x = date,
    y = Web_Design,
    name = 'Web Design (UX/UI)',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        dash = 'dot')
)

data = [trace0, trace1, trace2, trace3, trace4]

# Edit the layout
layout = dict(title = 'The Overall-Score Trend of Top5 Category Bootcamp',
              xaxis = dict(title = 'Year'),
              yaxis = dict(title = 'Score'),
              )

fig = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig)
```

```python
number_of_bootcamp = df.groupby('category').size().sort_values(ascending=False)
total_bootcamp = sum(number_of_bootcamp)
number_of_bootcamp = pd.DataFrame(number_of_bootcamp)
number_of_bootcamp.columns =['number']
number_of_bootcamp['porpotion'] = number_of_bootcamp.number/total_bootcamp
top_8 = number_of_bootcamp[:8]
all_review =combine.groupby('category').size().sort_values(ascending=False)
all_review= pd.DataFrame(all_review)
top_8_names = top_8.index.values.tolist()
all_review.columns =['number']
top_8_reviews = all_review.loc[top_8_names]
```

```python
y_saving = top_8.porpotion.values[::-1]*100
y_net_worth = top_8_reviews.number.values[::-1]
x_bootcamp = top_8.index[::-1]
x_net_worth = top_8.index[::-1]

trace0 = go.Bar(
    x=y_saving,
    y=x_bootcamp,
    marker=dict(
        color='rgba(50, 171, 96, 0.6)',
        line=dict(
            color='rgba(50, 171, 96, 1.0)',
            width=1),
    ),
    name='The Number of Bootcamps in Different Categories',
    orientation='h',
)
trace1 = go.Scatter(
    x=y_net_worth,
    y=x_net_worth,
    mode='lines+markers',
    line=dict(
        color='rgb(128, 0, 128)'),
    name='The number of Reviews in Different Categories',
)
layout = dict(
    title='Most Popular Bootcamps Categories',
    yaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=True,
        domain=[0, 0.85],
    ),
    yaxis2=dict(
        showgrid=False,
        showline=True,
        showticklabels=False,
        linecolor='rgba(102, 102, 102, 0.8)',
        linewidth=2,
        domain=[0, 0.85],
    ),
    xaxis=dict(
        zeroline=False,
        showline=False,
        showticklabels=True,
        showgrid=True,
        domain=[0, 0.42],
    ),
    xaxis2=dict(
        zeroline=False,
        showline=False,
        showticklabels=True,
        showgrid=True,
        domain=[0.47, 1],
        side='top',
        dtick=1000,
    ),
    legend=dict(
        x=0.029,
        y=1.038,
        font=dict(
            size=15,
        ),
    ),
    margin=dict(
        l=200,
        r=10,
        t=70,
        b=70,
    ),
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
)

annotations = []

y_s = np.round(y_saving, decimals=2)
y_nw = y_net_worth

# Adding labels
for ydn, yd, xd in zip(y_nw, y_s, x_bootcamp):
    # labeling the scatter savings
    annotations.append(dict(xref='x2', yref='y2',
                            y=xd, x=ydn,
                            text=ydn,
                            xanchor='right',
                            yanchor='bottom',
                            font=dict(family='Arial', size=12,
                                      color='rgb(128, 0, 128)'),
                            showarrow=False))
    # labeling the bar net worth
    annotations.append(dict(xref='x1', yref='y1',
                            y=xd, x=yd + 3,
                            text=str(yd) + '%',
                            font=dict(family='Arial', size=15,
                                      color='rgb(50, 171, 96)'),
                            showarrow=False))

layout['annotations'] = annotations

# Creating two subplots
fig = tools.make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=True,
                          shared_yaxes=False, vertical_spacing=0.001)

fig.append_trace(trace0, 1, 1)
fig.append_trace(trace1, 1, 2)

fig['layout'].update(layout)
plotly.offline.iplot(fig)
```

```python
analyser = SentimentIntensityAnalyzer()
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score
review_available = review[~(review.review_content =='')]
review_available = review_available.reset_index()
review_available['review_content_score'] = review.review_content.apply(sentiment_analyzer_scores)
review_available['negative_review'] = review_available.review_content_score.apply(lambda x: 1 if x['neg']>x['pos'] else 0)
review_available['low_score'] = review_available.overall_score.apply(lambda x : 1 if x <90 else 0)
review_available['negative_feedback'] = np.logical_or(review_available['negative_review']==1, review_available["low_score"]==1)
review_combine = pd.merge(df,review_available,how = 'inner',on = 'bootcamp_name')
```
